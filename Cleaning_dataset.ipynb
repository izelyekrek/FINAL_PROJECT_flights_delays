# Cleaning the dataset

## Importing the libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
pd.options.display.max_columns = None
```

## Importing the dateset 

```python
data = pd.read_csv('Dataset/uncleaned_dataset.csv')
data
```

## Normalizing my dataset

```python
data.columns = data.columns.str.lower()
data.columns = [column.lower().replace(' ', '_') for column in data.columns]
data
```

## Dropping the first column

```python
data = data.drop(['unnamed:_0'], axis = 1)
data
```

## Checking for null values and duplicates

```python
data.isna().sum()
```

```python
data = data.dropna()
```

```python
data1 = data.drop_duplicates()
data1
```

## Checking the outliers and removing them if necessary 

```python
sns.boxplot(x=data['age'])
plt.show()
```

```python
sns.boxplot(x=data['flight_distance'])
plt.show()
```

```python
iqr = np.percentile(data["flight_distance"],75) - np.percentile(data["flight_distance"],25) 
iqr
```

```python
upper_limit = np.percentile(data["flight_distance"],75) + 1.5*iqr
lower_limit = np.percentile(data["flight_distance"],25) - 1.5*iqr
upper_limit, lower_limit
```

```python
data = data[(data["flight_distance"]>lower_limit) & (data["flight_distance"]<upper_limit)]
data
```

```python
sns.boxplot(x=data['flight_distance'])
plt.show()
```
